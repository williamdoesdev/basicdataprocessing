```python
pd.options.display.float_format = '{:.0f}'.format
# pd.reset_option('display.float_format')

import re

columns = ['port_of_unlading', 'month']
for i in range(1, 100):
    columns.append("hs_" + str(i))
hs_by_unlading = pd.DataFrame(columns=columns)

header = pd.read_csv('datalake/layer=silver/table=header/year=2018/header_part0.csv')
tariff = pd.read_csv('datalake/layer=silver/table=tariff/year=2018/tariff_part0.csv')
tariff2 = pd.read_csv('datalake/layer=silver/table=tariff/year=2018/tariff_part1.csv')

header['harmonized_number'] = header['harmonized_number'].apply(
                    lambda x: int(re.sub('\D', '', str(x)[0:2])))

header['actual_arrival_date'] = header['actual_arrival_date'].apply(
                    lambda x: re.sub('-', '', re.search('-\d*-', x).group()))
header.rename(columns={'actual_arrival_date': 'arrival_month'}, inplace=True)


unlading = header.groupby(['arrival_month', 'port_of_unlading'])['harmonized_number'].apply(list).reset_index()

# new_columns = pd.DataFrame(columns=unlading['harmonized_number'].to_list())

for i in range(1, 100):
    unlading.append("hs_" + str(i))

def maphs(x):
    for i, element in x:
        column_name = "hs_" + str(element)
        unlading[column_name] = unlading[column_name] + 1
        break
    return

unlading['harmonized_number'].to_list().map(maphs)

unlading.head(50)
```
